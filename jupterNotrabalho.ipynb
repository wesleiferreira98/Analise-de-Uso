# Importação do data set a ser usado


```python
import pandas as pd
import zipfile
import gdown
import os



# URL do arquivo no Google Drive
drive_url = "https://drive.google.com/uc?id=1mHErmi7eyo2b6an8tCOsj1FoJ54N7Pkg"

# Caminho para salvar o arquivo
zip_path = "UCI_HAR_Dataset.zip"

# Baixar o arquivo usando gdown
gdown.download(drive_url, zip_path, quiet=False)

# Extração do arquivo zip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("UCI_HAR_Dataset")

# Caminho dos dados extraídos
data_path = "UCI_HAR_Dataset/UCI HAR Dataset"

# Listar todos os arquivos e diretórios dentro da pasta extraída
for root, dirs, files in os.walk("UCI_HAR_Dataset"):
    print(f"Diretório: {root}")
    for file in files:
        print(f"  Arquivo: {file}")

# Etapa 2: Carregamento dos Dados
features = pd.read_csv("UCI_HAR_Dataset/UCI HAR Dataset/features.txt", delim_whitespace=True, header=None)
X_train = pd.read_csv("UCI_HAR_Dataset/UCI HAR Dataset/train/X_train.txt", delim_whitespace=True, header=None)
y_train = pd.read_csv("UCI_HAR_Dataset/UCI HAR Dataset/train/y_train.txt", header=None)
```

# Implementação da  Deep Network Neural (DNN)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential  # type: ignore
from tensorflow.keras.layers import Dense, Dropout # type: ignore
from tensorflow.keras.optimizers import Adam # type: ignore
import matplotlib.pyplot as plt


# Etapa 1: Normalização dos dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)  

# Etapa 2: Seleção de features com SelectKBest
selector = SelectKBest(score_func=mutual_info_classif, k=50)
X_selected = selector.fit_transform(X_scaled, y_train.values.ravel())

# Etapa 3: Divisão dos dados em treino e teste
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(
    X_selected, y_train.values.ravel(), test_size=0.2, random_state=42, stratify=y_train
)

# Ajustando os rótulos para começar de 0, se necessário
y_train_split = y_train_split - 1
y_test_split = y_test_split - 1


# Etapa 4: Construção do modelo DNN
model = Sequential([
    Dense(128, activation='relu', input_dim=X_train_split.shape[1]),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(len(set(y_train_split)), activation='softmax')  # Saída para cada classe
])

# Compilação do modelo
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Etapa 5: Treinamento
history = model.fit(X_train_split, y_train_split, 
                    validation_split=0.2, epochs=30, batch_size=32, verbose=1)

# Etapa 6: Avaliação no conjunto de teste
test_loss, test_accuracy = model.evaluate(X_test_split, y_test_split, verbose=0)

# Previsões e métricas
y_pred = model.predict(X_test_split)
y_pred_classes = y_pred.argmax(axis=1)

print(f"Accuracy no conjunto de teste: {test_accuracy}")
print("Relatório de Classificação:")
print(classification_report(y_test_split, y_pred_classes))
print("Matriz de Confusão:")
print(confusion_matrix(y_test_split, y_pred_classes))

# Etapa 7: Visualização do histórico de treinamento
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Acurácia Treino')
plt.plot(history.history['val_accuracy'], label='Acurácia Validação')
plt.title('Acurácia do Modelo')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Perda Treino')
plt.plot(history.history['val_loss'], label='Perda Validação')
plt.title('Perda do Modelo')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.show()
```

# Implementação com as Redes de Função Base Radial (RBF - Radial Basis Function)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from sklearn.linear_model import Ridge
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Número de neurônios na camada oculta
n_centers = 200  

# Etapa 1: Encontrar os centros usando K-means
kmeans = KMeans(n_clusters=n_centers, random_state=42)
kmeans.fit(X_train_split)
centers = kmeans.cluster_centers_

# Etapa 2: Calcular a matriz RBF (usando a função Gaussiana)
def rbf_kernel(X, centers, gamma):
    return np.exp(-gamma * pairwise_distances(X, centers) ** 2)

# Calcular a matriz RBF para o treino e teste
gamma = 0.5  # Parâmetro de suavização da RBF
X_train_rbf = rbf_kernel(X_train_split, centers, gamma)
X_test_rbf = rbf_kernel(X_test_split, centers, gamma)

# Codificar os rótulos em formato one-hot
encoder = OneHotEncoder(sparse_output=False)
y_train_encoded = encoder.fit_transform(y_train_split.reshape(-1, 1))

# Ajustar o modelo Ridge com os rótulos one-hot
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_rbf, y_train_encoded)

# Fazer previsões
y_pred_rbf = ridge.predict(X_test_rbf)

# Converter previsões one-hot de volta para rótulos originais
y_pred_classes = np.argmax(y_pred_rbf, axis=1)

# Avaliar o desempenho
from sklearn.metrics import classification_report, confusion_matrix

print("Relatório de Classificação:")
print(classification_report(y_test_split, y_pred_classes))

print("Matriz de Confusão:")
print(confusion_matrix(y_test_split, y_pred_classes))
```

# Implementação com uma MLP

```python
from tensorflow.keras.models import Sequential # type: ignore
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization # type: ignore
from tensorflow.keras.optimizers import Adam # type: ignore
from sklearn.metrics import classification_report, confusion_matrix

# Construção do modelo MLP
model = Sequential([
    Dense(128, activation='relu', input_dim=X_train_rbf.shape[1]),  # Camada de entrada
    BatchNormalization(),
    Dropout(0.3),  # Regularização
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(len(set(y_train_split)), activation='softmax')  # Saída com softmax para múltiplas classes
])

# Compilação do modelo
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',  # Multiclasse
              metrics=['accuracy'])

# Treinamento
history = model.fit(X_train_rbf, y_train_split, 
                    validation_split=0.2, epochs=30, batch_size=32, verbose=1)

# Avaliação no conjunto de teste
test_loss, test_accuracy = model.evaluate(X_test_rbf, y_test_split, verbose=0)

# Previsões
y_pred = model.predict(X_test_rbf)
y_pred_classes = np.argmax(y_pred, axis=1)

# Relatório de Classificação
print("Relatório de Classificação:")
print(classification_report(y_test_split, y_pred_classes))

# Matriz de Confusão
print("Matriz de Confusão:")
print(confusion_matrix(y_test_split, y_pred_classes))

# Visualização do histórico de treinamento
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Acurácia Treino')
plt.plot(history.history['val_accuracy'], label='Acurácia Validação')
plt.title('Acurácia do Modelo')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Perda Treino')
plt.plot(history.history['val_loss'], label='Perda Validação')
plt.title('Perda do Modelo')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.show()
```

# Implementação com Rede Neural SOM (Self-Organizing Map)

```python
from sklearn.preprocessing import OneHotEncoder
from minisom import MiniSom
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

# Etapa 1: Inicialização e treinamento do SOM
som_size = 50  # Tamanho da grade do SOM
som = MiniSom(som_size, som_size, X_train_split.shape[1], sigma=1.0, learning_rate=0.5, random_seed=42)

# Treinamento do SOM com os dados de treino
som.train_random(X_train_split, 100)

# Etapa 2: Mapear as amostras para a rede SOM
# Obter os rótulos das células para as amostras de treino
train_labels = np.array([som.winner(x) for x in X_train_split])

# Mapear as amostras de teste para as células do SOM
test_labels = np.array([som.winner(x) for x in X_test_split])

# Etapa 3: Classificação usando a célula vencedora
# Codificar os rótulos em formato one-hot para o SOM
encoder = OneHotEncoder(sparse_output=False)
y_train_encoded = encoder.fit_transform(y_train_split.reshape(-1, 1))
y_test_encoded = encoder.transform(y_test_split.reshape(-1, 1))

# Etapa 4: Avaliar o desempenho
# A célula vencedora de cada amostra será associada ao seu rótulo original
y_pred_classes = []
for label in test_labels:
    # A célula vencedora é mapeada para o rótulo correspondente
    x, y = label
    closest_class = y_train_encoded[np.argmin(np.linalg.norm(X_train_split - som.get_weights()[x, y], axis=1))]
    y_pred_classes.append(np.argmax(closest_class))

# Avaliar o desempenho
print("Relatório de Classificação:")
print(classification_report(y_test_split, y_pred_classes))

print("Matriz de Confusão:")
print(confusion_matrix(y_test_split, y_pred_classes))

# Visualização do SOM (opcional)
plt.figure(figsize=(10, 8))
plt.imshow(som.distance_map().T, cmap='coolwarm')  # Mapa de distâncias
plt.colorbar()
plt.title("Mapa Auto-Organizável (SOM)")
plt.show()
```

# Implementação RBF e Rede neurais RAM

```python
from sklearn.decomposition import PCA
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import pairwise_distances

# Função RBF
def rbf_kernel(X, centers, gamma=0.5):
    return np.exp(-gamma * pairwise_distances(X, centers) ** 2)

# Redução de Dimensionalidade (PCA) para ajustar a entrada à memória
pca = PCA(n_components=50)  # Reduzindo para 50 componentes
X_train_rbf_reduced = pca.fit_transform(X_train_rbf)  # Ajustar a entrada
X_test_rbf_reduced = pca.transform(X_test_rbf)  # Ajustar a entrada de teste

# Definindo a classe de rede RAM com memória externa
class MemoryNetwork(nn.Module):
    def __init__(self, input_dim, mem_size, output_dim):
        super(MemoryNetwork, self).__init__()
        self.memory = nn.Parameter(torch.randn(mem_size, input_dim))  # Memória externa
        self.memory_size = mem_size
        self.input_dim = input_dim
        self.output_dim = output_dim
        
        # Rede de acesso à memória
        self.fc = nn.Linear(input_dim, output_dim)
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        # Cálculo da similaridade entre a entrada e a memória
        similarities = torch.matmul(self.memory, x.T)  # Similaridade entre memória e entrada
        attention_weights = self.softmax(similarities)  # Atenção para os dados mais relevantes

        # A soma ponderada da memória com base nos pesos de atenção
        output = torch.matmul(attention_weights.T, self.memory)
        return self.fc(output)  # Retorno final após passagem pela camada linear

# Parâmetros do modelo
input_dim = 50  # Número de características após redução
mem_size = 500  # Número de centros de memória
output_dim = len(set(y_train_split))  # Número de classes

# Inicializar a rede RAM
model = MemoryNetwork(input_dim, mem_size, output_dim)

# Definir critério de perda e otimizador
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Codificar os rótulos em formato one-hot
encoder = OneHotEncoder(sparse_output=False)
y_train_encoded = encoder.fit_transform(y_train_split.reshape(-1, 1))
y_test_encoded = encoder.transform(y_test_split.reshape(-1, 1))

# Convertendo as entradas para Tensor
X_train_tensor = torch.tensor(X_train_rbf_reduced, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_rbf_reduced, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_split, dtype=torch.long)
y_test_tensor = torch.tensor(y_test_split, dtype=torch.long)

# Treinamento
num_epochs = 150
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()

    # Previsão
    output = model(X_train_tensor)
    
    # Cálculo da perda
    loss = criterion(output, y_train_tensor)
    
    # Backpropagation
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Avaliação no conjunto de teste
model.eval()
with torch.no_grad():
    output = model(X_test_tensor)
    _, predicted = torch.max(output, 1)

    # Relatório de Classificação
    print("Relatório de Classificação:")
    print(classification_report(y_test_split, predicted.numpy()))

    # Matriz de Confusão
    print("Matriz de Confusão:")
    print(confusion_matrix(y_test_split, predicted.numpy()))
```
